```python

from googleapiclient.discovery import build
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import re 
from unidecode import unidecode 

try:
    nltk.download('vader_lexicon')
except Exception as e:
    print(f"Pesan saat mencoba mengunduh vader_lexicon (mungkin sudah ada atau ada error): {e}")
    print("Jika ada error LookupError setelah ini, pastikan NLTK dapat mengunduh resource.")



API_KEY = 'AIzaSyCS7xPhcgKBAQIvsv3eUdUkvikC-LuA3Tw'  
VIDEO_ID = 'sa4FbbS8FzA'  


KATA_INDIKASI_JUDOL = [
    "777","77","88", "gacor", "slot gacor", "maxwin", "big win", "mega win", "super win","jepe","gajir","toto",
    "jackpot (jp)", "auto jp", "auto cuan", "winrate tinggi", "mudah menang", 
    "hoki parah", "hoki gila", "hoki sejati", "spin hoki", "turbo spin", 
    "buyspin hoki", "buyspin receh", "buyspin langsung jp", "scatter gacor", 
    "banjir scatter", "full scatter", "room gacor", "room sultan", "room sakti", 
    "room dewa", "trik gacor", "trik anti zonk", "settingan admin", 
    "settingan sakti", "settingan gacor", "modal receh, hasil sultan", 
    "modal 10k tembus 1jt", "receh jadi sultan", "cuan terus", "cuan gila", 
    "cuan nonstop", "cuan maksimal", "langsung cuan", "langsung profit", 
    "langsung maxwin", "langsung wd", "langsung kaya", "auto withdraw", 
    "auto win", "auto hoki", "auto sultan", "sultan mode", "sultan room", 
    "sultan spin", "badai cuan", "badai jp", "badai scatter", "ledakan jp", 
    "ledakan cuan", "petir bersinar", "petir cuan", "petir jp", "sensasional", 
    "gila roomnya", "gila cuan", "pecah scatter", "pecah jp", "tembus maxwin", 
    "tembus jackpot", "top win", "top player", "cair", "sultan player", 
    "jitu", "jitu hoki", "anti loss", "anti rungkad", "no rungkad", "bebas loss", 
    "pola gacor", "pola hari ini", "slot gacorr", "slot 77", "jp jam segini", 
    "slot hari ini", "jam hoki", "modal kecil profit", "dari 10k ke juta", 
    "transferan masuk", "isi saldo", "wd kilat", "wd cepat", "room jp", 
    "slot terbaik", "slot online", "slot anti rungkad", "petir menyambar", 
    "hujan cuan", "scatter melimpah", "bocoran slot", "pattern win", 
    "cuan slot", "admin settingan", "bot slot", "bot jp", "top cuan", 
    "sultan jp", "gila menang", "slot auto jp", "trik modal receh", 
    "auto profit", "cu4n", "sl0t", "s|ot", "slothoki", "cuan harian", 
    "viral slot", "trending slot", "modal 10k", "maxw1n", "makwin", "mw"
]

KATA_INDIKASI_JUDOL_LOWER = [kata.lower() for kata in KATA_INDIKASI_JUDOL]



youtube = build('youtube', 'v3', developerKey=API_KEY)


try:
    sid = SentimentIntensityAnalyzer()
except LookupError:
    print("LookupError saat inisialisasi SentimentIntensityAnalyzer. Pastikan vader_lexicon sudah terunduh.")
    print("Anda bisa mencoba menjalankan sel di atas untuk mengunduh vader_lexicon lagi.")

    sid = None 



def get_video_details(video_id_input):
    """Mengambil judul dan ID kategori video."""
    try:
        request = youtube.videos().list(
            part="snippet,statistics", 
            id=video_id_input
        )
        response = request.execute()
        if response['items']:
            snippet = response['items'][0]['snippet']
            video_title = snippet['title']
            category_id = snippet['categoryId']
            
            cat_request = youtube.videoCategories().list(
                part="snippet",
                id=category_id
            )
            cat_response = cat_request.execute()
            video_category_title = "Unknown"
            if cat_response['items']:
                video_category_title = cat_response['items'][0]['snippet']['title']
            
            return video_title, video_category_title
        else:
            return "Video Not Found", "Unknown"
    except Exception as e:
        print(f"Error saat mengambil detail video: {e}")
        return "Error Fetching Title", "Error Fetching Category"


def analyze_comment_text(comment_text_original): 
    """Menganalisis teks komentar untuk kata indikasi judol dan sentimen."""
    

    normalized_comment_text = unidecode(comment_text_original) 

    comment_text_lower = normalized_comment_text.lower()

    
    ditemukan = []
    for kata_indikasi in KATA_INDIKASI_JUDOL_LOWER:

        if kata_indikasi in comment_text_lower:

            ditemukan.append(kata_indikasi) 

    jumlah_kata_indikasi = len(set(ditemukan)) 
    
    sent_score = 0.0
    sent_label = "Neutral (Sentiment Analyzer Error)"

    if sid:

        sentiment_scores = sid.polarity_scores(comment_text_original) 
        sent_score = sentiment_scores['compound']
        
        if sent_score >= 0.05:
            sent_label = "Positive"
        elif sent_score <= -0.05:
            sent_label = "Negative"
        else:
            sent_label = "Neutral"
            
    return list(set(ditemukan)), jumlah_kata_indikasi, sent_score, sent_label


def get_video_comments(video_id_input):

    
    video_title, video_category = get_video_details(video_id_input)
    if video_title == "Video Not Found" or video_title == "Error Fetching Title":
        print(f"Tidak bisa melanjutkan karena detail video {video_id_input} tidak ditemukan atau error.")
        return []

    all_comments_data = []
    next_page_token = None
    
    print(f"Mengambil komentar untuk video: '{video_title}' (Kategori: {video_category})")

    try:
        while True:
            request = youtube.commentThreads().list(
                part='snippet,replies', 
                videoId=video_id_input,
                maxResults=100,  
                textFormat='plainText',
                pageToken=next_page_token
            )
            response = request.execute()

            for item in response['items']:
                top_level_comment_snippet = item['snippet']['topLevelComment']['snippet']
                
                comment_text = top_level_comment_snippet['textDisplay'] 
                author_display_name = top_level_comment_snippet['authorDisplayName']
                author_channel_id_value = top_level_comment_snippet.get('authorChannelId', {}).get('value', 'N/A')
                published_at = top_level_comment_snippet['publishedAt']
                like_count = top_level_comment_snippet['likeCount']
                reply_count = item['snippet']['totalReplyCount']
                
                comment_length_char = len(comment_text) 
                
      
                kata_ditemukan, jumlah_kata, sent_score, sent_label = analyze_comment_text(comment_text) 
                
                is_judol = 1 if jumlah_kata > 0 else 0
                
                all_comments_data.append({
                    'video_id': video_id_input,
                    'video_title': video_title,
                    'video_category': video_category,
                    'comment_text': comment_text,
                    'author_display_name': author_display_name,
                    'author_channel_id': author_channel_id_value,
                    'published_at': published_at,
                    'like_count': like_count,
                    'reply_count': reply_count,
                    'comment_length_char': comment_length_char,
                    'kata_indikasi_judol_ditemukan': ", ".join(kata_ditemukan) if kata_ditemukan else "",
                    'jumlah_kata_indikasi_judol': jumlah_kata,
                    'sentiment_score': sent_score,
                    'sentiment_label': sent_label,
                    'is_judol_comment': is_judol
                })
            
            next_page_token = response.get('nextPageToken')
            if not next_page_token:
                break 
            print(f"Mengambil halaman komentar berikutnya...")

        print(f"Selesai mengambil {len(all_comments_data)} komentar.")
        return all_comments_data

    except Exception as e:
        print(f"Terjadi error saat mengambil komentar: {e}")
        return all_comments_data


if API_KEY == 'MASUKKAN_API_KEY_ANDA_DI_SINI':
    print("!!! PERHATIAN: Harap ganti 'MASUKKAN_API_KEY_ANDA_DI_SINI' dengan API Key YouTube Anda yang valid. !!!")
elif not API_KEY or len(API_KEY) < 30: 
     print("!!! PERHATIAN: API Key YouTube Anda tampaknya tidak valid atau kosong. Harap periksa kembali. !!!")
else:

    comments_data = get_video_comments(VIDEO_ID)

    if comments_data:
 
        df_comments = pd.DataFrame(comments_data)
        
        print("\nContoh Data Komentar:")
        print(df_comments.head())
        
        print("\nInfo DataFrame:")
        df_comments.info()
        

        nama_file_excel = f"komentar_{VIDEO_ID.replace('-', '_')}.xlsx" 
        try:
            df_comments.to_excel(nama_file_excel, index=False, engine='openpyxl') 
            print(f"\nData telah disimpan ke {nama_file_excel}")
        except Exception as e:
            print(f"\nError saat menyimpan ke Excel: {e}")
            print("Pastikan Anda memiliki library 'openpyxl' terinstal: pip install openpyxl")

    else:
        print("Tidak ada data komentar yang berhasil diproses.")
```
