```python
# %% [markdown]
# # Analisis Komentar YouTube
# Notebook ini bertujuan untuk mengambil komentar dari video YouTube, melakukan analisis konten untuk mendeteksi kata-kata indikasi "judol" (judi online), dan melakukan analisis sentimen dasar.
# 
# **Kolom yang akan dihasilkan:**
# 1.  `video_id`
# 2.  `video_title`
# 3.  `video_category`
# 4.  `comment_text`
# 5.  `author_display_name`
# 6.  `author_channel_id`
# 7.  `published_at`
# 8.  `like_count`
# 9.  `reply_count`
# 10. `comment_length_char`
# 11. `kata_indikasi_judol_ditemukan`
# 12. `jumlah_kata_indikasi_judol`
# 13. `sentiment_score`
# 14. `sentiment_label`
# 15. `is_judol_comment`

# %% [python]
# Import library yang diperlukan
from googleapiclient.discovery import build
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import re # Untuk pembersihan teks dasar jika diperlukan

# Download VADER lexicon jika belum (cukup sekali)
try:
    nltk.data.find('sentiment/vader_lexicon.zip')
except nltk.downloader.DownloadError:
    nltk.download('vader_lexicon')

# %% [python]
# --- KONFIGURASI ---
API_KEY = 'AIzaSyCS7xPhcgKBAQIvsv3eUdUkvikC-LuA3Tw'  # Ganti dengan API Key YouTube Anda
VIDEO_ID = 'sa4FbbS8FzA'  # Ganti dengan ID video yang ingin dianalisis

# Daftar 120 kata/frasa indikasi "judol"
# (Disarankan untuk menormalkan ke huruf kecil untuk pencocokan case-insensitive)
KATA_INDIKASI_JUDOL = [
    "777", "gacor", "slot gacor", "maxwin", "big win", "mega win", "super win","dewa","jepe", 
    "jackpot (jp)", "auto jp", "auto cuan", "winrate tinggi", "mudah menang", 
    "hoki parah", "hoki gila", "hoki sejati", "spin hoki", "turbo spin", 
    "buyspin hoki", "buyspin receh", "buyspin langsung jp", "scatter gacor", 
    "banjir scatter", "full scatter", "room gacor", "room sultan", "room sakti", 
    "room dewa", "trik gacor", "trik anti zonk", "settingan admin", 
    "settingan sakti", "settingan gacor", "modal receh, hasil sultan", 
    "modal 10k tembus 1jt", "receh jadi sultan", "cuan terus", "cuan gila", 
    "cuan nonstop", "cuan maksimal", "langsung cuan", "langsung profit", 
    "langsung maxwin", "langsung wd", "langsung kaya", "auto withdraw", 
    "auto win", "auto hoki", "auto sultan", "sultan mode", "sultan room", 
    "sultan spin", "badai cuan", "badai jp", "badai scatter", "ledakan jp", 
    "ledakan cuan", "petir bersinar", "petir cuan", "petir jp", "sensasional", 
    "gila roomnya", "gila cuan", "pecah scatter", "pecah jp", "tembus maxwin", 
    "tembus jackpot", "top win", "top player", "cair", "sultan player", 
    "jitu", "jitu hoki", "anti loss", "anti rungkad", "no rungkad", "bebas loss", 
    "pola gacor", "pola hari ini", "slot gacorr", "slot 77", "jp jam segini", 
    "slot hari ini", "jam hoki", "modal kecil profit", "dari 10k ke juta", 
    "transferan masuk", "isi saldo", "wd kilat", "wd cepat", "room jp", 
    "slot terbaik", "slot online", "slot anti rungkad", "petir menyambar", 
    "hujan cuan", "scatter melimpah", "bocoran slot", "pattern win", 
    "cuan slot", "admin settingan", "bot slot", "bot jp", "top cuan", 
    "sultan jp", "gila menang", "slot auto jp", "trik modal receh", 
    "auto profit", "cu4n", "sl0t", "s|ot", "slothoki", "cuan harian", 
    "viral slot", "trending slot", "modal 10k", "maxw1n", "makwin", "mw"
]
# Normalisasi kata indikasi ke huruf kecil
KATA_INDIKASI_JUDOL_LOWER = [kata.lower() for kata in KATA_INDIKASI_JUDOL]


# %% [python]
# Inisialisasi YouTube API client
youtube = build('youtube', 'v3', developerKey=API_KEY)

# Inisialisasi Sentiment Analyzer (VADER)
sid = SentimentIntensityAnalyzer()

# %% [python]
def get_video_details(video_id_input):
    """Mengambil judul dan ID kategori video."""
    try:
        request = youtube.videos().list(
            part="snippet,statistics", # Menambahkan statistics untuk info lain jika perlu
            id=video_id_input
        )
        response = request.execute()
        if response['items']:
            snippet = response['items'][0]['snippet']
            video_title = snippet['title']
            category_id = snippet['categoryId']
            
            # Mendapatkan nama kategori dari ID
            cat_request = youtube.videoCategories().list(
                part="snippet",
                id=category_id
            )
            cat_response = cat_request.execute()
            video_category_title = "Unknown"
            if cat_response['items']:
                video_category_title = cat_response['items'][0]['snippet']['title']
            
            return video_title, video_category_title
        else:
            return "Video Not Found", "Unknown"
    except Exception as e:
        print(f"Error saat mengambil detail video: {e}")
        return "Error Fetching Title", "Error Fetching Category"

# %% [python]
def analyze_comment_text(comment_text):
    """Menganalisis teks komentar untuk kata indikasi judol dan sentimen."""
    comment_text_lower = comment_text.lower()
    
    # Deteksi kata indikasi judol
    ditemukan = []
    for kata_indikasi in KATA_INDIKASI_JUDOL_LOWER:
        # Menggunakan \b untuk word boundary agar tidak match substring (misal "slot" di "slots")
        # Perlu diperhatikan jika kata indikasi mengandung karakter spesial regex
        # Untuk kesederhanaan, kita pakai 'in' dulu, tapi regex lebih robas
        # Contoh regex: pattern = r'\b' + re.escape(kata_indikasi) + r'\b'
        if kata_indikasi in comment_text_lower:
            # Cari kata asli (dengan case) jika memungkinkan, atau gunakan versi lower
            # Ini bisa kompleks, untuk sekarang kita pakai versi lower saja
            original_case_found = False
            for original_kata_indikasi in KATA_INDIKASI_JUDOL:
                 if original_kata_indikasi.lower() == kata_indikasi and original_kata_indikasi in comment_text:
                      ditemukan.append(original_kata_indikasi)
                      original_case_found = True
                      break
            if not original_case_found:
                 ditemukan.append(kata_indikasi) # fallback ke versi lower jika case asli tidak ketemu persis

    jumlah_kata_indikasi = len(set(ditemukan)) # Gunakan set untuk hitungan unik jika ada duplikasi kata indikasi
    
    # Analisis Sentimen menggunakan VADER
    # Catatan: VADER lebih optimal untuk bahasa Inggris. Hasil untuk Bahasa Indonesia mungkin bervariasi.
    sentiment_scores = sid.polarity_scores(comment_text)
    sentiment_score = sentiment_scores['compound'] # Skor compound: -1 (negatif) hingga +1 (positif)
    
    sentiment_label = "Neutral"
    if sentiment_score >= 0.05:
        sentiment_label = "Positive"
    elif sentiment_score <= -0.05:
        sentiment_label = "Negative"
        
    return list(set(ditemukan)), jumlah_kata_indikasi, sentiment_score, sentiment_label

# %% [python]
def get_video_comments(video_id_input):
    """Mengambil dan memproses komentar dari video."""
    
    video_title, video_category = get_video_details(video_id_input)
    if video_title == "Video Not Found" or video_title == "Error Fetching Title":
        print(f"Tidak bisa melanjutkan karena detail video {video_id_input} tidak ditemukan atau error.")
        return []

    all_comments_data = []
    next_page_token = None
    
    print(f"Mengambil komentar untuk video: '{video_title}' (Kategori: {video_category})")

    try:
        while True:
            request = youtube.commentThreads().list(
                part='snippet,replies', # replies untuk totalReplyCount
                videoId=video_id_input,
                maxResults=100, 
                textFormat='plainText',
                pageToken=next_page_token
            )
            response = request.execute()

            for item in response['items']:
                top_level_comment_snippet = item['snippet']['topLevelComment']['snippet']
                
                comment_text = top_level_comment_snippet['textDisplay']
                author_display_name = top_level_comment_snippet['authorDisplayName']
                # Handle jika authorChannelId tidak ada atau formatnya berbeda
                author_channel_id_value = top_level_comment_snippet.get('authorChannelId', {}).get('value', 'N/A')
                published_at = top_level_comment_snippet['publishedAt']
                like_count = top_level_comment_snippet['likeCount']
                reply_count = item['snippet']['totalReplyCount'] # Jumlah balasan untuk top-level comment ini
                
                comment_length_char = len(comment_text)
                
                kata_ditemukan, jumlah_kata, sent_score, sent_label = analyze_comment_text(comment_text)
                
                is_judol = 1 if jumlah_kata > 0 else 0
                
                all_comments_data.append({
                    'video_id': video_id_input,
                    'video_title': video_title,
                    'video_category': video_category,
                    'comment_text': comment_text,
                    'author_display_name': author_display_name,
                    'author_channel_id': author_channel_id_value,
                    'published_at': published_at,
                    'like_count': like_count,
                    'reply_count': reply_count,
                    'comment_length_char': comment_length_char,
                    'kata_indikasi_judol_ditemukan': ", ".join(kata_ditemukan) if kata_ditemukan else "",
                    'jumlah_kata_indikasi_judol': jumlah_kata,
                    'sentiment_score': sent_score,
                    'sentiment_label': sent_label,
                    'is_judol_comment': is_judol
                })
            
            next_page_token = response.get('nextPageToken')
            if not next_page_token:
                break # Keluar dari loop jika tidak ada halaman berikutnya
            print(f"Mengambil halaman komentar berikutnya...")

        print(f"Selesai mengambil {len(all_comments_data)} komentar.")
        return all_comments_data

    except Exception as e:
        print(f"Terjadi error saat mengambil komentar: {e}")
        return all_comments_data # Kembalikan data yang sudah terkumpul jika ada error

# %% [python]
# --- EKSEKUSI UTAMA ---
if API_KEY == 'MASUKKAN_API_KEY_ANDA_DI_SINI':
    print("!!! PERHATIAN: Harap ganti 'MASUKKAN_API_KEY_ANDA_DI_SINI' dengan API Key YouTube Anda yang valid. !!!")
else:
    # Ambil dan proses komentar
    comments_data = get_video_comments(VIDEO_ID)

    if comments_data:
        # Buat DataFrame Pandas
        df_comments = pd.DataFrame(comments_data)
        
        # Tampilkan beberapa baris pertama DataFrame
        print("\nContoh Data Komentar:")
        print(df_comments.head())
        
        # Tampilkan info DataFrame
        print("\nInfo DataFrame:")
        df_comments.info()
        
        # (Opsional) Simpan ke file Excel
        nama_file_excel = f"komentar_{VIDEO_ID}.xlsx"
        df_comments.to_excel(nama_file_excel, index=False)
        print(f"\nData telah disimpan ke {nama_file_excel}")
    else:
        print("Tidak ada data komentar yang berhasil diproses.")

# %% [markdown]
# ---
# ### Catatan Tambahan:
# * **Paginasi Komentar**: Kode di atas sudah diupdate untuk menangani paginasi (mengambil lebih dari 100 komentar jika ada).
# * **Kata Indikasi Judol**: Pencocokan kata indikasi saat ini menggunakan `in` pada string yang sudah di-lowercase. Untuk pencocokan yang lebih akurat (misalnya, hanya kata utuh, bukan substring), Anda bisa menggunakan regular expressions (`re.search(r'\b' + re.escape(kata_indikasi) + r'\b', comment_text_lower)`).
# * **Analisis Sentimen (VADER)**: Seperti yang disebutkan, VADER dioptimalkan untuk bahasa Inggris. Untuk Bahasa Indonesia, pertimbangkan library atau model lain seperti `indonesia-bert-sentiment` atau melatih model sendiri jika akurasi tinggi sangat krusial.
# * **Error Handling**: Kode ini menyertakan beberapa `try-except` dasar. Anda mungkin ingin menambahkannya lebih lanjut tergantung kebutuhan.
# * **Kategori Video**: Kode mengambil ID kategori dan kemudian mencoba mengambil nama kategori. Ini melibatkan panggilan API tambahan.
# * **`author_channel_id`**: Penanganan ditambahkan jika `authorChannelId` tidak ada atau formatnya berbeda.
# ---

```
